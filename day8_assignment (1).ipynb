```python
def armstrong():
    for i in range(1, 1000 + 1):
        order = len(str(i))
        s = 0
        temp = i
        while temp > 0:
            digit = temp % 10
            s += digit ** order
            temp //= 10
        if i==s:
            yield i
```

```python
for j in armstrong():
        print(j)
```
